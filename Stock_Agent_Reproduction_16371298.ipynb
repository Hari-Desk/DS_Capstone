```python
import re

def patch_util(debug=True):
    util_file = "util.py"
    with open(util_file, "r") as f:
        code = f.read()

    if debug:
        print("üîß Switching to DEBUG mode (fast simulation)...")
        code = re.sub(r"AGENTS_NUM\s*=\s*\d+", "AGENTS_NUM = 3", code)
        code = re.sub(r"TOTAL_DATE\s*=\s*\d+", "TOTAL_DATE = 5", code)
        code = re.sub(r"TOTAL_SESSION\s*=\s*\d+", "TOTAL_SESSION = 1", code)
        code = re.sub(r"MAX_INITIAL_PROPERTY\s*=\s*\d+\.?\d*", "MAX_INITIAL_PROPERTY = 500000.0", code)
        code = re.sub(r"MIN_INITIAL_PROPERTY\s*=\s*\d+\.?\d*", "MIN_INITIAL_PROPERTY = 10000.0", code)
    else:
        print("üöÄ Switching to FULL mode (full-scale simulation)...")
        code = re.sub(r"AGENTS_NUM\s*=\s*\d+", "AGENTS_NUM = 50", code)
        code = re.sub(r"TOTAL_DATE\s*=\s*\d+", "TOTAL_DATE = 264", code)
        code = re.sub(r"TOTAL_SESSION\s*=\s*\d+", "TOTAL_SESSION = 3", code)
        code = re.sub(r"MAX_INITIAL_PROPERTY\s*=\s*\d+\.?\d*", "MAX_INITIAL_PROPERTY = 5000000.0", code)
        code = re.sub(r"MIN_INITIAL_PROPERTY\s*=\s*\d+\.?\d*", "MIN_INITIAL_PROPERTY = 100000.0", code)

    with open(util_file, "w") as f:
        f.write(code)
    print("‚úÖ util.py patched successfully!")

# --- USAGE ---
# Debug mode (quick runs)
patch_util(debug=True)

# Full mode (real experiment)
# patch_util(debug=False)
```
